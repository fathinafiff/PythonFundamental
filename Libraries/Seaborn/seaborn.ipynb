```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

## Scatterplot

```python
"""
hue -> grouping
size -> size of dots
palette -> color
"""
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size="size", palette="YlGnBu")
```

## Histogram

```python
"""
kde -> kernel density estimation
bins -> number of bins
"""
sns.histplot(tips['tip'], kde=True, bins=15)
```

```python
sns.displot(tips['tip'], kde=True, bins=15)
```

## Barplot

```python
"""
hue -> grouping
palette -> color
"""
sns.barplot(x='sex', y='tip', data=tips, hue='sex', palette='YlGnBu')
```

## Box Plot

```python
"""
hue -> grouping
palette -> color
"""
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex', palette='YlGnBu')
```

## Strip Plot

```python
"""
dodge -> grouping
"""
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu', dodge=True)
```

## Join Plot

```python
"""
kind -> change
"""
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
"""
kind -> change the type of plot
fill -> fill the plot
cmap -> color
"""
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde', fill=True, cmap='YlGnBu')
```

```python
"""
kind -> change the type of plot
cmap -> color
"""
sns.jointplot(x='tip', y='total_bill', data=tips, kind='hex', cmap='YlGnBu')
```

## Pair Plot

```python
titanic.head()
```

```python
"""
titanic.select_dtypes(['number']) -> select only numeric columns
"""
sns.pairplot(titanic.select_dtypes(['number']), hue='pclass', palette='YlGnBu')
```

## Heat Map

```python
titanic.corr()
```

```python
"""
annot -> show the value
cmap -> color [icefire, coolwarm, YlGnBu, etc.]
"""
sns.heatmap(data=titanic.corr(), annot=True, cmap='icefire')
```
