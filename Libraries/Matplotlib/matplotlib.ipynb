```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

# Basic Useful Graph

## Scatter Plot

```python
# Make random data
X_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100
```

```python
# Scatter Plot
"""
c or color -> Use for color, ex: Blue, Red, or HexColor
marker -> Style of dots
s -> Size of dots
alpha -> transparency
"""
plt.scatter(X_data, y_data, color='blue', marker='*', s=150, alpha=0.3)
plt.show()
```

## Line Chart

```python
years = [2006 + x for x in range(16)]
weights = [int(round(23 + x*10,0)) for x in np.random.random(16)]
```

```python
"""
lw -> Line Weight
linestyle -> Style of the line
"""
plt.plot(years, weights, color='green', lw=3, linestyle="--")
plt.show()
```

## Bar chart

```python
x = ["C++", "C#", "Python", "Java", "GO"]
y = [20, 50, 100, 1 , 46]
```

```python
"""
align -> for aligning the bar chart
width -> Size of the bar chart
edgecolor -> color of bar edges 
"""

plt.bar(x, y, color='r', align='edge', width=0.2, edgecolor='g', lw=6)
plt.show()
```

## Histogram

```python
ages = np.random.normal(20, 1.5, 1000)
```

```python
"""
bins -> Specify number of bins
"""
plt.hist(ages, color='red', bins=[ages.min(), 18, 21, ages.max()])
plt.show()
```

#### Cumulative histogram

```python
"""
cumulative -> [True/False] Default False
"""
plt.hist(ages, color='red', cumulative=True)
plt.show()
```

## Pie Chart

```python
x = ["C++", "C#", "Python", "Java", "GO"]
y = [20, 50, 100, 1 , 46]
```

```python
"""
explode -> Separate the chart a little bit from other {Using same position as label}
autopct -> Add percentage into graph
pctdistance -> Move percentage distance from center
startangle -> Move angle of pie chart
"""
plt.pie(y, labels=x, explode=[0, 0, 0.3, 0, 0], autopct='%.2f%%', pctdistance=0.5, startangle=90)
plt.show()
```

## Box Plot

```python
heights = np.random.normal(172, 8, 390)
```

```python
plt.boxplot(heights)
plt.show()
```

# Plot Customization

## Adding Title and Label

```python
years = [int(f'20{x}') for x in range(14,22)]
income = [53, 52, 60, 53, 63, 72, 78, 73]
```

```python
income_ticks = list(range(50,81,2))
```

```python
plt.plot(years, income)
plt.title('Income of Jhon (in USD)', fontsize=20, fontname="Arial")
plt.xlabel('Year')
plt.ylabel('Yearly Income in USD')
plt.yticks(income_ticks, [f"{x}k USD" for x in income_ticks]) # Help make y ticks
plt.show()
```

# Using Legend

```python
AMD = [11, 12, 13, 14, 15]
NVDA = [101, 102, 103, 104, 105]
BA = [21, 22, 23, 24, 25]
```

```python
"""
loc -> for aligning the legend
"""

plt.plot(AMD)
plt.plot(NVDA)
plt.plot(BA)
plt.legend(['AMD', 'NVDA', 'BA'], loc='upper right')
plt.show()
```

## Style Sheets

```python
"""
https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
https://matplotlib.org/stable/users/explain/customizing.html
"""
votes = [10, 2, 16, 34]
people = ['A', 'B', 'C', 'D']

style.use('ggplot')

plt.pie(votes, labels=people)
plt.legend(people)
plt.show()
```

## Multiple Figures

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)
```

```python
plt.figure(1)
plt.scatter(x1, y1)
plt.figure(2)
plt.plot(x2, y2)

plt.show()
```

### Subplots

```python
x = np.arange(100)
style.use('ggplot')

fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].plot(x, np.sin(x), color='b')
ax[0, 0].set_title('sin(x)')

ax[0, 1].plot(x, np.tan(x), c='g')
ax[0, 1].set_title('tan(x)')

ax[1, 0].plot(x, np.random.random(100), c='r')
ax[1, 0].set_title('Random')

ax[1, 1].plot(x, np.log(x), c='k')
ax[1, 1].set_title('Log(x)')

fig.suptitle("Four Plot")
plt.show()
```

## Save Plot

```python
"""
tight_layout -> Remove white space
    pad -> Padding
    w_pad -> Padding width
    h_pad -> Padding height
savefig -> Save image
    dpi -> Resolution
    transparent -> Remove background
    bbox_inches -> Remove white space
"""

plt.tight_layout(pad=1, w_pad=1, h_pad=1)
plt.savefig('fourplots.png', dpi=300, transparent=True, bbox_inches='tight')
```

# 3 Dimension Plot

```python
x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax = plt.axes(projection='3d')

ax.scatter(x, y, z)
ax.set_title('3D Plot')
ax.set_xlabel('test')
plt.show()
```

```python
x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)

ax = plt.axes(projection='3d')

ax.plot(x, y, z)
ax.set_title('3D Plot')
ax.set_xlabel('test')
plt.show()
```

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)

Z = np.sin(X) + np.cos(Y)

ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap = 'Spectral')
ax.set_title('3D Plot')
ax.set_xlabel('test')
ax.view_init(azim=30, elev=45)
plt.show()
```

# Animating Plots

```python
import random

head_tails = [0, 0]

for _ in range(5):
    head_tails[random.randint(0, 1)] += 1
    plt.bar(['Heads', 'Tails'], head_tails, color=['b', 'r'])
    plt.pause(0.001)
plt.show()
```
